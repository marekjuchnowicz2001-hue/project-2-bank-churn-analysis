3. Overall Target Analysis

In this step we analyze the target variable of the dataset.

The goal is to understand the overall churn rate and the distribution of the target variable.

We want to check:

-how many customers stayed

-how many customers left the bank

-the overall churn rate

```python
df["Exited"].value_counts()
```

```python
df["Exited"].value_counts(normalize=True)
```

```python
churn_rate = df["Exited"].mean()
churn_rate
```

 4. Feature vs Target Analysis

In this step we analyze how different customer features relate to churn.

The goal is to identify patterns and potential factors influencing customer attrition.

```python
df.groupby("IsActiveMember")["Exited"].mean()
```

```python
df.groupby("Geography")["Exited"].mean()
```

```python
df.groupby("Gender")["Exited"].mean()
```

```python
df.groupby("NumOfProducts")["Exited"].mean()
```

```python
df.groupby("Exited")["Age"].mean()
```

```python
df.groupby("Exited")["Balance"].mean()
```

Key Insights

-The overall customer churn rate is approximately 20%, indicating a significant customer retention challenge.

-Inactive customers are nearly twice as likely to churn compared to active customers, highlighting the importance of customer engagement.

-Customers located in Germany show the highest churn rate, suggesting potential regional differences in competition or service satisfaction.

-Female customers exhibit higher churn rates than male customers.

-Customers holding two bank products are the most stable group, while customers with three or more products show extremely high churn rates.

-Customers who leave the bank tend to be older and hold higher account balances, indicating that the bank may be losing high-value clients.
